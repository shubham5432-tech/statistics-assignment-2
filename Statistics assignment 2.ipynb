### 1. What is a random variable in probability theory?

A **random variable** is a numerical value assigned to the outcome of a random experiment.

It acts like a function that maps outcomes of a sample space to real numbers.

There are two main types:
- **Discrete random variable**: Takes specific, countable values (e.g., number of heads in coin tosses).
- **Continuous random variable**: Can take any value in a continuous range (e.g., height, weight).

#### Example:
- Tossing a coin twice:
  - Possible outcomes: HH, HT, TH, TT
  - Define a random variable X = number of heads
  - Then X can be 0, 1, or 2 depending on the result.
---
### 2. What are the types of random variables?

There are two main types of random variables:

#### 1. **Discrete Random Variable**  
- Takes on a **finite or countably infinite** number of values.  
- Common in situations where outcomes are counted.  
- Example: Number of students in a class, number of heads in 3 coin tosses.

#### 2. **Continuous Random Variable**  
- Can take on **any value within a continuous range**.

---

#### üìå Key Difference:

### 4. What are Probability Distribution Functions (PDF)?

#### üìå Types of PDFs:

1. **For Discrete Random Variables**:  
   It's called a **Probability Mass Function (PMF)**.  
   - It gives the probability for each **exact** value.
   - Example: Rolling a fair die ‚Äî each face (1‚Äì6) has P = 1/6.

2. **For Continuous Random Variables**:  
   It‚Äôs called a **Probability Density Function (PDF)**.  
   - It gives a **density**, not direct probability.
   - Probability is calculated over an **interval**, not a single point.
   - The **area under the curve** between two points gives the probability.

---

#### ‚úèÔ∏è Key Properties:
- For a continuous PDF, total area under the curve = 1.
- A PDF **is not a probability**, but the integral (area) over a range is.
---
### 5. How do Cumulative Distribution Functions (CDF) differ from Probability Distribution Functions (PDF)?

#### üìò PDF (Probability Distribution Function)
- Applies to **continuous** random variables.
- Describes the **density** of probabilities over a range.
- Cannot be used directly to find the probability at a single point (it's zero for continuous variables).
- To find probability over a range, integrate the PDF.

#### üìò CDF (Cumulative Distribution Function)
- Applies to **both discrete and continuous** variables.
- Represents the **probability that a random variable X is less than or equal to a value x**:  
  \[
  CDF(x) = P(X \leq x)
  \]
- Always increases from 0 to 1.
- For continuous variables, CDF is the **area under the PDF curve** from -‚àû to x.

---

### üîÑ Key Differences:

| Feature              | PDF                                        | CDF                                        |
|---------------------|--------------------------------------------|--------------------------------------------|
| Meaning             | Probability **density** at a point         | Cumulative probability up to a point       |
| Value Range         | Can be greater than 1 (as density)         | Always between 0 and 1                     |
| Area or Height      | Height of the curve                        | Area under the curve up to x               |
| Type                | Only for continuous distributions          | For both continuous and discrete           |

### 6. What is a Discrete Uniform Distribution?

### 7. What are the Key Properties of a Bernoulli Distribution?

A **Bernoulli distribution** is a discrete probability distribution for a random experiment that has only **two possible outcomes**:
- Success (1) with probability **p**
- Failure (0) with probability **1 - p**

---

### ‚úÖ Properties:

- **Random Variable X**: Takes value 1 (success) or 0 (failure)
- **Probability Mass Function (PMF)**:
  \[
  P(X = x) = p^x (1 - p)^{1 - x}, \quad \text{for } x \in \{0,1\}
  \]
- **Mean (Expected Value)**:  
  \[
  E[X] = p
  \]
- **Variance**:  
  \[
  Var(X) = p(1 - p)
  \]

---

### üß† Example:
- Tossing a coin:
  - Success = Head (1), Failure = Tail (0)
  - If coin is fair, then p = 0.5
---

### 8. What is the Binomial Distribution, and How Is It Used in Probability?

The **Binomial Distribution** models the number of **successes** in a fixed number of **independent Bernoulli trials**, where each trial has two outcomes (success or failure).

---

### ‚úÖ Key Parameters:
- **n**: Number of trials
- **p**: Probability of success in each trial
- **X**: Random variable representing the number of successes (X ‚àà {0, 1, ..., n})

---

### üìò Probability Mass Function (PMF):
$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$
Where:
 $$
 ( \binom{n}{k} \) = "n choose k" = number of ways to choose k successes
 $$

---

### üìä Common Use Cases:
- Number of heads in 10 coin tosses
- Number of correct answers in a multiple-choice quiz
- Number of defective items in a batch
---
### 9. What is the Poisson Distribution and Where Is It Applied?

The **Poisson distribution** is a discrete probability distribution that models the number of **events occurring in a fixed interval** of time or space, **given the average number of events (Œª)** and assuming the events occur independently.

---

### ‚úÖ Key Parameter:
- **Œª (lambda)**: Average rate (mean) of occurrences per interval (e.g., 5 calls per hour)

---

### üìò Probability Mass Function (PMF):

$$
P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}
$$

Where:
- **k** = number of events (0, 1, 2, ...)
- **Œª** = average rate
- **e** ‚âà 2.71828

---

### üìå Common Applications:
- Number of customer arrivals per hour at a store
- Number of emails received in a day
- Number of defects per unit area in manufacturing
- Earthquakes in a region per year

---

### üìä Characteristics:
- Discrete, non-negative integer values
- Mean = Variance = Œª
- Assumes events happen **independently** and **randomly**

### 10. What is a Continuous Uniform Distribution?

A **continuous uniform distribution** is a type of probability distribution where **every value within a specified continuous interval [a, b] is equally likely**.

### üìò Probability Density Function (PDF):

### 11. What Are the Characteristics of a Normal Distribution?

### 12. What is the Standard Normal Distribution, and Why Is It Important?

The **Standard Normal Distribution** is a special case of the normal distribution where:

- **Mean (Œº) = 0**
- **Standard deviation (œÉ) = 1**

It is denoted by **Z**, and the random variable that follows this distribution is called a **Z-score**.

---

### ‚úÖ Why Is It Important?

   - Any normal distribution can be converted into a standard normal using Z-scores:
     \[
     Z = \frac{X - \mu}{\sigma}
     \]

2. **Simplifies Calculations**  
   - Tables and calculators use Z-distribution to compute probabilities and critical values.

3. **Used in Hypothesis Testing**  
   - Z-values are critical in determining significance in statistical tests.

4. **Foundation for Confidence Intervals**  
   - Z-distribution helps define boundaries for confidence levels like 90%, 95%, and 99%.

5. **Common in CLT Applications**  
   - Due to the Central Limit Theorem, sample means approach a normal distribution ‚Üí often standardized.

### 13. What is the Central Limit Theorem (CLT), and Why Is It Critical in Statistics?

The **Central Limit Theorem (CLT)** is a fundamental concept in statistics that explains how the distribution of sample means becomes **approximately normal**, regardless of the shape of the original population distribution, **as the sample size becomes large**.

---

### ‚úÖ Formal Statement:
If you take many random samples (size **n**) from any population (with mean Œº and standard deviation œÉ), then:

- The distribution of the **sample means** approaches a **normal distribution**
- The mean of the sampling distribution is **Œº**
- The standard deviation of the sampling distribution is **œÉ / ‚àön**

This happens when **n is sufficiently large** (usually n ‚â• 30).

---

### üìå Why Is It Critical?

1. **Enables Normal Approximation**  
   - Even if the population is not normally distributed, sample means will be approximately normal.

2. **Foundation of Inferential Statistics**  
   - Used to create confidence intervals and conduct hypothesis tests.

3. **Allows Use of Z-Statistics**  
   - Because sampling distribution becomes normal, we can use standard normal tables for analysis.

4. **Real-World Applications**  
   - Estimating averages like exam scores, customer spending, wait times, etc.

---

### üß† Example:

Suppose we measure the height of 100 different samples of students, each of size 50. Even if individual heights are skewed, the **average height of each sample** will form a **normal distribution**.

---
### 14. How Does the Central Limit Theorem Relate to the Normal Distribution?

The **Central Limit Theorem (CLT)** explains **why** the **normal distribution is so commonly used** in statistics ‚Äî even when the original data is **not normally distributed**.

---

### ‚úÖ Key Relationship:

- As per CLT, when you:
  - Take many **random samples** (with size n)
  - From **any population** (regardless of its distribution)

$$

### üìå Why This Matters:

1. **Justifies the Use of Z-scores and Confidence Intervals**

   - Works even when the original population is skewed or uniform.

3. **Forms the Basis of Hypothesis Testing**  
   - Especially for tests involving means and proportions.

### 15. What is the Application of Z Statistics in Hypothesis Testing?

**Z-statistics** (or **Z-scores**) are used in **hypothesis testing** to determine whether a sample result is significantly different from a known population parameter, assuming the data follows a normal distribution.

---

### ‚úÖ Common Applications:

1. **Testing a Population Mean**  
  $$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$

Where:  

$$
\bar{X} : \text{ Sample mean}
$$

$$
\mu : \text{ Population mean}
$$

$$
\sigma : \text{ Population standard deviation}
$$

$$
n : \text{ Sample size}
$$

2. **Comparing Sample and Population**  
   To check whether a sample is likely drawn from a population with a known mean.

3. **Used in One-Tailed or Two-Tailed Tests**  
   - **One-tailed**: Directional claim (e.g., is mean > Œº?)  
   - **Two-tailed**: Non-directional claim (e.g., is mean ‚â† Œº?)

---

### üìå Decision Rule:

- Calculate the **Z-statistic** using sample data.
- Compare it with a **critical Z-value** (e.g., ¬±1.96 for 95% confidence).
- If it falls in the rejection region, **reject the null hypothesis (H‚ÇÄ)**.

---
### 16. How Do You Calculate a Z-Score, and What Does It Represent?

A **Z-score** tells you **how many standard deviations** a data point is from the **mean** of the distribution.

---

### ‚úÖ Z-Score Formula:

For a **single data point** \( X \):

$$
Z = \frac{X - \mu}{\sigma}
$$

For a **sample mean** \( \bar{X} \) (when sampling distribution is used):

$$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$

---

### üß† What It Represents:

- **Z = 0** ‚Üí The value is exactly at the mean  
- **Z > 0** ‚Üí The value is above the mean  
- **Z < 0** ‚Üí The value is below the mean  
- It helps compare values **from different distributions** or **detect outliers**

---

### üìä Practical Use:

Z-scores are used in:
- **Standardization** of datasets
- **Hypothesis testing**
- **Confidence interval** calculation
- **Detecting outliers** (e.g., if \( |Z| > 2 \), the point may be unusual)

---
### 17. What Are Point Estimates and Interval Estimates in Statistics?

In statistics, we often use sample data to **estimate population parameters** (like mean or proportion). There are two main types of estimates:

---

### ‚úÖ 1. Point Estimate

- A **single value** used to estimate a population parameter.
- Example: Using the **sample mean ( \( \bar{X} \) )** to estimate the population mean \( \mu \).

#### üìå Example:
> If you measure the average height of 30 students and get 165 cm,  
> then **165 cm is the point estimate** of the population mean height.

---

### ‚úÖ 2. Interval Estimate (Confidence Interval)

- Provides a **range of values** likely to contain the population parameter.
- More informative than a point estimate because it includes **uncertainty**.
- Usually written as:

$$
\text{Confidence Interval} = \bar{X} \pm Z \cdot \frac{\sigma}{\sqrt{n}}
$$

Where:

$$
\bar{X} = \text{Sample mean}
$$

$$
Z = \text{Z-score for the desired confidence level (e.g., 1.96 for 95%)}
$$

$$
\sigma = \text{Population standard deviation}
$$

$$
n = \text{Sample size}
$$

### 18. What is the Significance of Confidence Intervals in Statistical Analysis?

A **confidence interval (CI)** is a range of values, derived from sample data, that is likely to contain the true value of an unknown **population parameter** (like mean or proportion) with a specified level of confidence (e.g., 95%).

---

### ‚úÖ Why Confidence Intervals Matter:

1. **Incorporate Uncertainty**  
   - CIs reflect the **natural variability** in sample data and give a **range** instead of a single point.

2. **More Informative Than Point Estimates**  
   - A point estimate (like a sample mean) is just a guess. A CI gives a **margin of error** around that guess.

3. **Allow Decision-Making with Risk Awareness**  
   - Helps researchers and policymakers understand **how precise** their estimates are.

4. **Used in Hypothesis Testing**  
   - If a hypothesized value (e.g., population mean) **falls outside the CI**, it may be **rejected**.

5. **Communicate Statistical Results Clearly**  
   - A 95% CI means that if we repeated the experiment many times, **95% of those intervals would contain the true parameter**.

---

### üìò Example:

- Suppose the average test score from a sample of students is 80, and the 95% confidence interval is (77.5, 82.5).  
- This means we're 95% confident the true average score of **all students** lies between **77.5 and 82.5**.

---

### 21. What Are the Assumptions for Applying the Central Limit Theorem (CLT)?

The **Central Limit Theorem (CLT)** states that the sampling distribution of the sample mean approaches a **normal distribution** as the sample size becomes large ‚Äî **regardless of the shape of the population**.

However, there are a few important **assumptions** to apply CLT correctly:

---

### ‚úÖ Key Assumptions:

1. **Random Sampling**  
   - The samples should be drawn **randomly** from the population.

2. **Independence**  
   - Observations must be **independent** of each other.  
   - If sampling without replacement, the sample size should be **less than 10% of the population** to ensure approximate independence.

3. **Sample Size is Sufficiently Large**  
   - If the population is **not normally distributed**, a **larger sample size** is needed (commonly **n ‚â• 30**).  
   - If the population **is normal**, smaller sample sizes are acceptable.

4. **Finite Variance**  
   - The population from which the samples are drawn must have a **finite standard deviation** (i.e., no infinite values or extreme outliers dominating the data).

---

### üß† Summary Table:

| Assumption              | Why It Matters                          |
|-------------------------|------------------------------------------|
| Random Sampling         | Avoids bias in estimation                |
| Independence            | Ensures valid distribution behavior      |
| Large Enough Sample     | Enables approximation to normality       |
| Finite Variance         | Keeps sampling distribution stable       |

---

### üìò Example:

- If you're drawing student test scores from a skewed population, you need **n ‚â• 30** to trust that the sample mean will behave normally.

### 22. What is the Concept of Expected Value in a Probability Distribution?

The **expected value** (or **mean**) of a random variable is the **long-run average outcome** you'd expect if an experiment were repeated many times.

---

### ‚úÖ Definition:

For a **discrete** random variable:

Let the values be:

$$
x_1, x_2, \ldots, x_n
$$

with corresponding probabilities:

$$
P(x_1), P(x_2), \ldots, P(x_n)
$$

Then the **expected value** is calculated as:

$$
E(X) = \sum_{i=1}^{n} x_i \cdot P(x_i)
$$

For a **continuous** random variable with probability density function \( f(x) \):

$$
E(X) = \int_{-\infty}^{\infty} x \cdot f(x) \, dx
$$

---

### üß† Interpretation:

- The expected value is **not necessarily an outcome** itself.
- It's a **weighted average** of all possible outcomes.
- Think of it as the **"center of mass"** of the distribution.

---

### üìò Example (Discrete):

> Suppose you roll a fair 6-sided die. The expected value is:

$$
E(X) = \frac{1+2+3+4+5+6}{6} = 3.5
$$

> You never actually roll a 3.5 ‚Äî but that's the average **in the long run**.

---
### 23. How Does a Probability Distribution Relate to the Expected Outcome of a Random Variable?

A **probability distribution** tells us the **likelihood** of each possible outcome of a **random variable**.

The **expected outcome** (or **expected value**) is the **weighted average** of all those possible outcomes, where each outcome is weighted by its probability from the distribution.

---

### ‚úÖ Relationship Explained:

- The **probability distribution** provides:
  - All possible values a random variable can take
  - The **probability** associated with each value

- The **expected value** is calculated using the distribution as:

$$
E(X) = \sum_{i=1}^{n} x_i \cdot P(x_i)
\quad \text{(discrete case)}
$$

or

$$
E(X) = \int_{-\infty}^{\infty} x \cdot f(x) \, dx
\quad \text{(continuous case)}
$$

---

### üìò Example:

> A loaded die gives these probabilities:
> - 1 with 0.1, 2 with 0.1, 3 with 0.2, 4 with 0.2, 5 with 0.2, 6 with 0.2  
> Then the **expected value** is:

$$
E(X) = 1(0.1) + 2(0.1) + 3(0.2) + 4(0.2) + 5(0.2) + 6(0.2) = 4.2
$$

This value comes directly from **applying the probability distribution** to the variable‚Äôs possible outcomes.

###3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): Outcome (must be 0 or 1)
    p (float): Probability of success (0 ‚â§ p ‚â§ 1)

    Returns:
    float: Probability of observing x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)

# Example usage
p = 0.7  # Probability of success

print("P(X=0):", bernoulli_pdf(0, p))  # Should return 0.3
print("P(X=1):", bernoulli_pdf(1, p))  # Should return 0.7
```

###6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1. Define a discrete uniform distribution (e.g., integers from 1 to 6)
# -------------------------------
values = np.arange(1, 7)  # Like rolling a fair 6-sided die
prob = 1 / len(values)    # Equal probability for each value

# -------------------------------
# 2. Calculate the Cumulative Distribution Function (CDF)
# -------------------------------
pmf = np.full_like(values, prob, dtype=float)
cdf = np.cumsum(pmf)  # Cumulative sum of PMF

# -------------------------------
# 3. Plot the CDF
# -------------------------------
plt.figure(figsize=(8, 5))
plt.step(values, cdf, where='post', color='darkblue', linewidth=2)
plt.title("CDF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.ylim(0, 1.1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(values)
plt.show()

# -------------------------------
# 4. Display CDF values
# -------------------------------
for v, c in zip(values, cdf):
    print(f"P(X ‚â§ {v}) = {c:.2f}")
```

###9.Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores and plot them for a given dataset.

    Parameters:
    data (list or np.array): Numeric dataset
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    z_scores = (data - mean) / std

    # Print values
    print("Mean:", round(mean, 2))
    print("Standard Deviation:", round(std, 2))
    print("Z-scores:", np.round(z_scores, 2))

    # Plot Z-scores
    plt.figure(figsize=(10, 5))
    plt.bar(range(len(z_scores)), z_scores, color='coral', edgecolor='black')
    plt.axhline(0, color='black', linestyle='--')
    plt.title("Z-scores of Dataset")
    plt.xlabel("Index")
    plt.ylabel("Z-score")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example usage
sample_data = [65, 70, 75, 80, 85, 90, 95, 100]
calculate_and_plot_z_scores(sample_data)
```

###14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mu, sigma):
    """
    Calculate Z-score for a given data point x.
    """
    z = (x - mu) / sigma
    return z

# -------------------------------
# Example values
# -------------------------------
x = 75           # data point
mu = 60          # population mean
sigma = 10       # population standard deviation

# -------------------------------
# Calculate Z-score
# -------------------------------
z = calculate_z_score(x, mu, sigma)
print(f"Z-score for x = {x} is: {z:.2f}")

# -------------------------------
# Plot Standard Normal Distribution with Z-score
# -------------------------------
x_vals = np.linspace(-4, 4, 1000)
y_vals = norm.pdf(x_vals, 0, 1)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='Standard Normal PDF', color='blue')

# Fill area up to the Z-score
plt.fill_between(x_vals, y_vals, where=(x_vals <= z), color='skyblue', alpha=0.5, label=f'P(Z ‚â§ {z:.2f})')

# Mark the Z-score
plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')

plt.title("Z-score Compared to Standard Normal Distribution")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

###15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

# -------------------------------
# 1. Sample data and population parameters
# -------------------------------
sample_data = [52, 48, 50, 47, 51, 49, 53, 46, 50, 49]
mu_0 = 50         # Hypothesized population mean (null)
sigma = 2         # Known population standard deviation
alpha = 0.05      # Significance level

# -------------------------------
# 2. Compute test statistic (Z)
# -------------------------------
sample_mean = np.mean(sample_data)
n = len(sample_data)

z = (sample_mean - mu_0) / (sigma / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))  # two-tailed test

# -------------------------------
# 3. Output the results
# -------------------------------
print("Sample Mean:", round(sample_mean, 2))
print("Z-Statistic:", round(z, 3))
print("P-value:", round(p_value, 4))

# -------------------------------
# 4. Decision
# -------------------------------
if p_value < alpha:
    print("‚ùå Reject the Null Hypothesis (Significant Difference)")
else:
    print("‚úÖ Fail to Reject the Null Hypothesis (No Significant Difference)")
```

###16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy.stats import norm

# -------------------------------
# 1. Sample data and parameters
# -------------------------------
data = [52, 48, 50, 47, 51, 49, 53, 46, 50, 49]
sigma = 2         # Known population standard deviation
confidence = 0.95 # Confidence level

# -------------------------------
# 2. Compute sample statistics
# -------------------------------
n = len(data)
sample_mean = np.mean(data)
z_score = norm.ppf(1 - (1 - confidence) / 2)

margin_of_error = z_score * (sigma / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# -------------------------------
# 3. Display result
# -------------------------------
print(f"Sample Mean = {sample_mean:.2f}")
print(f"{int(confidence * 100)}% Confidence Interval: [{ci_lower:.2f}, {ci_upper:.2f}]")

# -------------------------------
# 4. Interpretation
# -------------------------------
print(f"\nInterpretation:")
print(f"We are {int(confidence * 100)}% confident that the true population mean lies between {ci_lower:.2f} and {ci_upper:.2f}.")
```

###17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# -------------------------------
# 1. Generate data from a normal distribution
# -------------------------------
mu = 100       # population mean
sigma = 15     # population standard deviation
n = 50         # sample size

np.random.seed(1)  # for reproducibility
sample = np.random.normal(loc=mu, scale=sigma, size=n)

# -------------------------------
# 2. Calculate sample statistics
# -------------------------------
sample_mean = np.mean(sample)
sample_std = sigma  # known population std deviation

# Confidence level
confidence = 0.95
z = norm.ppf(1 - (1 - confidence) / 2)

# Margin of error
margin_of_error = z * (sample_std / np.sqrt(n))

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# -------------------------------
# 3. Print results
# -------------------------------
print(f"Sample Mean: {sample_mean:.2f}")
print(f"{int(confidence * 100)}% Confidence Interval: [{ci_lower:.2f}, {ci_upper:.2f}]")

# -------------------------------
# 4. Interpretation
# -------------------------------
print(f"\nInterpretation:")
print(f"We are {int(confidence * 100)}% confident that the true population mean lies between {ci_lower:.2f} and {ci_upper:.2f}.")

# -------------------------------
# 5. Optional: Plot the CI
# -------------------------------
plt.figure(figsize=(8, 2))
plt.axvline(mu, color='green', linestyle='--', label='True Mean (Œº)')
plt.errorbar(sample_mean, 0.5, xerr=margin_of_error, fmt='o', color='red', capsize=5, label='Confidence Interval')
plt.title("Confidence Interval for Sample Mean")
plt.yticks([])
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

###24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution_skew_kurt(mu=0, sigma=1, size=1000):
    """
    Generates data from a normal distribution and calculates skewness and kurtosis.

    Parameters:
        mu (float): Mean of the distribution
        sigma (float): Standard deviation
        size (int): Number of data points to generate

    Returns:
        dict: Contains skewness and kurtosis values
    """
    # Generate data
    data = np.random.normal(loc=mu, scale=sigma, size=size)

    # Calculate skewness and kurtosis
    data_skew = skew(data)
    data_kurtosis = kurtosis(data, fisher=True)  # Fisher=True gives 0 for normal

    # Display results
    print(f"Mean: {np.mean(data):.2f}, Std Dev: {np.std(data):.2f}")
    print(f"Skewness: {data_skew:.4f} (‚âà 0 means symmetric)")
    print(f"Kurtosis: {data_kurtosis:.4f} (‚âà 0 means normal-like)")

    return {"skewness": data_skew, "kurtosis": data_kurtosis}

# -------------------------------
# Call the function
# -------------------------------
analyze_normal_distribution_skew_kurt(mu=100, sigma=15, size=1000)
```
